```python
#YTTc1 is the 3year treasury bond future, YTCc1= is the 10y Treasury bond future
rics = ['YTTc1=', 'YTCc1=']
```

```python
#what is fields = '*'
Table = eikon.get_timeseries (rics,fields ='*',start_date = '2014-01-01', end_date = '2018-12-31', interval = 'daily',calendar='native')
```

```python
Table = Table.rename(columns={'YTTc1=':'a','YTCc1=':'b'})
# a=2y treasury bond future, b=10y treasury bond future
Table
```

```python
#diff() finds difference between list 1 and list 2
#set() convert any of the iterable to the distinct element and sorted sequence of iterable elements 
#.symmetric difference method returns a set that contains all items from both set, but not the items that are present in both sets
def diff(list1, list2):
    return list(set(list1).symmetric_difference(set(list2))) 
```

```python
#.index.tolist() convert the index to a list
Dates = Table.index.tolist()
a = Table['a']['CLOSE'].tolist()
b = Table['b']['CLOSE'].tolist()

curve =pd.DataFrame({'Dates':Dates,'a':a,'b':b})

#finding the 10y treasury and 3y treasury spread
curve['curve'] = curve['b'] -curve['a']

print(curve)

```

```python
spread = curve.loc[:,['Dates','curve']]
spread = spread.set_index(['Dates'])
spread.index = pd.to_datetime(spread.index)
```

```python
spread['curve+30'] = spread['curve'].shift(-1)
spread['diff'] = spread['curve+30'] - spread['curve']
spread['diff'] = spread['diff']
```
